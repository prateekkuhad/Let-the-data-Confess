```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the Dataset
df=pd.read_csv(r'googleplaystores.csv')
```

```python
df.info()
```

```python
# if any NA values are present, drop that row or column.
df.dropna()
```

```python
# replacing the empty positions
df.fillna(0,inplace=True)
```

```python
# checks of null values
df.isnull().sum()
```

```python
# Replacing K with *1000 <- Return it in float 

# Replacing M with *1000000 <- Return it in float 
def change(x):
    if type(x) == object or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0

df['Size'] = df['Size'].apply(change)
```

# 1) What is the range of the size of most of the apps?
  ###  Ans: 100 MB

```python
Smax=df['Size'].max()
print("Max range of apps is:",Smax)
```

```python
# Replacing the dollar '$' sign with space ' ' to avoid error
df['Price']=df['Price'].str.replace('$','')

# Converting price column values to Numeric
df['Price'] = pd.to_numeric(df['Price'],errors = 'coerce')
```

```python
# Find the maximum Price Available 
df['Price'].max()
```

# 3)What is the app name having max price?
### Ans : $400  I'm Rich - Trump Edition

```python
CApp=df['App'].loc[df['Price']==400.0]
print(CApp)
```

```python
# Replacing the plus '+' sign with space ' ' to avoid error
df['Installs']=df['Installs'].str.replace('+','')

# Replacing the comma ',' sign with space ' ' to avoid error
df['Installs'] = df['Installs'].str.replace(',','')
```

```python
df = df.drop(10472)
# Converting Installs which is String to Numeric
df['Installs'] = pd.to_numeric(df['Installs'])
```

```python
# Creating a separate dataframe for GAME Category 

a=df.loc[df['Category'] =='GAME']
a
```

```python
# Top 20 Apps with respect to no. of Installs in the Game Category ( including duplicates)

LApp=a.nlargest(20, ['Installs'])
print(LApp)
```

```python
#List of APPS in the game category
x1=a['App'].loc[a['Category']=='GAME']
x1
```

```python
# Apps in the Game Category which have a minimum of 500000000 downloads

x1=a['App'].loc[a['Installs']>=500000000]
print("Top 5 Apps based on Installation\n:",x1.unique())
```

```python
# Apps in the Game Category which have a minimum of 100000000 downloads

x2=a['App'].loc[a['Installs']==100000000]
print(x2.unique())
```

```python
# Selecting a 15 random apps 

random_selected = x2.sample(n=15)
print(random_selected.unique())
```

```python
# Storing the Unique values in variables

a1=(x1.unique())
a2=(random_selected.unique())

```

# 4)List top 20 apps in Game Category(Based on Number of Installation)

```python
# Combining the variables using np.concatenate

frames = [a1,a2]
result = np.concatenate(frames)
display(result)
```
